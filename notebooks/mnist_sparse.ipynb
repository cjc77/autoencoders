```python
import numpy as np
from keras.layers import Dense, Activation, LeakyReLU
from keras.models import Sequential
from keras.losses import MeanSquaredError, BinaryCrossentropy
from keras.regularizers import l1
from keras.datasets import mnist
```

```python
# (X_train, y_train), (X_test, y_test) = boston_housing.load_data()
(X_train, y_train), (X_test, y_test) = mnist.load_data()
X_train = np.reshape(X_train, (len(X_train), -1))
X_test = np.reshape(X_test, (len(X_test), -1))

X_train = X_train / 255.0
X_test = X_test / 255.0
d = X_train.shape[1]
# enc_d = 32
enc_d = d

print(f"{d}, {enc_d}")
```

```python
encoder = Sequential()
encoder.add(Dense(enc_d, input_dim=d, activation="relu", activity_regularizer=l1(10e-6)))

decoder = Sequential()
decoder.add(Dense(d, activation="sigmoid", input_dim=enc_d))

auto_enc = Sequential()
auto_enc.add(encoder)
auto_enc.add(decoder)
auto_enc.compile(optimizer="adam", loss=BinaryCrossentropy(), metrics=["accuracy"])
```

```python
encoder.summary()
decoder.summary()
auto_enc.summary()
```

```python
hist = auto_enc.fit(X_train, X_train, batch_size=64, epochs=20, validation_data=(X_test, X_test))
```
